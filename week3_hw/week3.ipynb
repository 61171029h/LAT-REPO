# 高級中等學校應屆畢業生未升學未就業情形(按縣市別分）

```python
import pandas as pd

df = pd.read_csv('109_1_2_7.csv')

df
```

先把比率刪掉留人數

```python
#:表示選擇所有行，'a':'b' 表示選擇從列標籤為 a 的列到列標籤為 b 的列之間的所有列。
temp = df.loc[:, '設立別':'其他人數']
temp
```

1. 各縣市畢業人數

```python
graduate = temp[['學校所在地別', '畢業生人數']]
graduate_grouped = graduate.groupby('學校所在地別').sum().reset_index()
graduate_grouped
```

2. 各縣市準備出國的比率(富有程度)

```python
#先算各縣市準備出國的人數
abroad = temp[['學校所在地別', '畢業生人數', '準備出國人數']]
abroad_grouped = abroad.groupby('學校所在地別').sum()
abroad_grouped
```

```python
#將畢業人數除準備出國人數
abroad_p = abroad_grouped['準備出國人數'].div(abroad_grouped['畢業生人數'])

#將畢業人數除準備出國人數結果換算成百分比
abroad_per = abroad_grouped.assign(畢業生準備出國比率=abroad_p.apply(lambda x: '{:.2%}'.format(x)))
abroad_per

#將內容依據出國比率由高至低排序
abroad_per_sorted = abroad_per.sort_values(by='畢業生準備出國比率', ascending=False)
abroad_per_sorted
```

 3. 各縣市接受職訓、找工作而未找到的比率(高中畢業即就業)

```python
#先算各縣市接受職訓、找工作而未找到的人數
work = temp[['學校所在地別', '畢業生人數','正在接受職業訓練人數', '需要工作而未找到人數']]
work_grouped = work.groupby('學校所在地別').sum()
work_grouped
```

```python
#將畢業人數除「接受職訓、找工作而未找到」人數
work_p= work_grouped['需要工作而未找到人數'].div(work_grouped['畢業生人數'])
train_p= work_grouped['正在接受職業訓練人數'].div(work_grouped['畢業生人數'])

#將畢業人數除「接受職訓、找工作而未找到」人數結果換算成百分比
work_per=work_p.apply(lambda x: '{:.2%}'.format(x))
train_per=train_p.apply(lambda x: '{:.2%}'.format(x))

#合併
work_grouped = work_grouped.assign(需要工作而未找到人數占比=work_per, 正在接受職業訓練人數占比=train_per)
work_grouped
```
