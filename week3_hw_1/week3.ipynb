```python
import pandas as pd

df = pd.read_csv('109_1_2_7.csv')

df
```

先把比率刪掉留人數

```python
temp = df.loc[:, '設立別':'其他人數']
temp
```

1. 各縣市畢業人數

```python
graduate = df[['學校所在地別', '畢業生人數']]
graduate
Type = graduate.groupby(['學校所在地別'])['畢業生人數'].sum().reset_index()
Type
```

2. 各縣市人數個別準備出國的比率

```python
#將縣市融合
abroad = df[['學校所在地別', '畢業生人數', '準備出國人數']]
abroad
```

```python
#縣市出國比率
abroad_p = abroad['準備出國人數'].div(abroad['畢業生人數'])
abroad_per = abroad.assign(畢業生準備出國比率=abroad_p.apply(lambda x: '{:.2%}'.format(x)))
abroad_per
```
