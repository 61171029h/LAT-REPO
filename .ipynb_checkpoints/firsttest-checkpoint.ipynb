# 高級中等學校應屆畢業生未升學未就業情形(按縣市別分）

```python
import pandas as pd

df = pd.read_csv('107_1_2_7.csv')

df.head()
```

```python
temp = df[['學校所在地別','需要工作而未找到比率']]
temp
```

```python
Type = temp.groupby(['學校所在地別'])['需要工作而未找到比率'].sum().reset_index()
```

```python
Type['設立別'] = '國立'
Type
```

```python
df107 = pd.read_csv('107_1_2_7.csv')
df108 = pd.read_csv('108_1_2_7.csv')
```

```python
new = df107[['學校所在地別', '需要工作而未找到比率']].groupby(['學校所在地別'])['需要工作而未找到比率'].sum().reset_index()
```

```python
new['設立別'] = '國立'
```

```python
new
```

```python
frames = [Type, new]

result = pd.concat(frames)
```
